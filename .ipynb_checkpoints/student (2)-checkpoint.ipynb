```python
from fuzzywuzzy import fuzz

# Function to find similar titles and drop duplicates
def drop_similar_titles(df, column, threshold=85):
    seen = set()  # To track processed titles
    unique_rows = []

    for index, row in df.iterrows():
        title = row[column]
        similar = False

        # Check against already seen titles
        for seen_title in seen:
            if fuzz.ratio(title, seen_title) > threshold:
                similar = True
                break
        
        # If not similar, add to unique rows and mark title as seen
        if not similar:
            unique_rows.append(row)
            seen.add(title)
    
    # Return a DataFrame with only unique rows
    return pd.DataFrame(unique_rows)
```

```python
# Example Usage
df = pd.DataFrame({'title': ['The Avengers', 'Avengers', 'Avenger', 'Iron Man', 'Iron Man 2']})
cleaned_df = drop_similar_titles(df, column='title')
print(cleaned_df)
```

## Repeat the process with for loop and functions 

```python
import pandas as pd

def merge_movie_datasets(df_1, df_2, df_3, df_4, df_5):
    """
    Merge multiple movie datasets and clean up duplicate columns.
    
    Args:
        df_1, df_2, df_3, df_4, df_5, df_6: Input dataframes
    Returns:
        Cleaned and merged dataframe
    """
    # Step 1: Initial merges
    # Title-based merge (Inner Merge to avoid duplicates)
    title_merge_1 = pd.merge(df_1, df_5, on='title', how='left')  # Change to 'inner'
    title_merge = pd.merge(title_merge_1, df_6, on='title', how='left')  # Change to 'inner'
    
    # ID-based merges (Inner Merge to avoid duplicates)
    id_merge = df_2.merge(df_3, on='id', how='left').merge(df_4, on='id', how='left')  # Change to 'inner'
    
    # Final merge (Inner Merge to avoid duplicates)
    merged_df = pd.merge(title_merge, id_merge, on='id', how='left')  # Change to 'inner'
    
    # Step 2: Clean up duplicate columns
    columns_to_combine = [
        'domestic_gross',
        'title',
        'studio',
        'release_date',
        'rating'
    ]
    
    # Create a copy to avoid SettingWithCopyWarning
    df = merged_df.copy()
    
    # Combine and clean up columns
    for col in columns_to_combine:
        x_col = f"{col}_x"
        y_col = f"{col}_y"
        
        # Only process if both columns exist
        if x_col in df.columns and y_col in df.columns:
            # Combine columns
            df[col] = df[x_col].combine_first(df[y_col])
            # Drop original columns
            df.drop(columns=[x_col, y_col], inplace=True, errors='ignore')
    
    # Step 3: Validation
    print("\nMerge Statistics:")
    print(f"Input shapes:")
    print(f"df_1: {df_1.shape}")
    print(f"df_2: {df_2.shape}")
    print(f"df_3: {df_3.shape}")
    print(f"df_4: {df_4.shape}")
    print(f"df_5: {df_5.shape}")
    print(f"df_6: {df_6.shape}")
    print(f"Final merged shape: {df.shape}")
    
    # Check for any remaining duplicate columns
    duplicate_cols = [col for col in df.columns if '_x' in col or '_y' in col]
    if duplicate_cols:
        print("\nWarning: Some duplicate columns remain:")
        print(duplicate_cols)
    
    return df

# Example usage:
cleaned_df = merge_movie_datasets(df_1, df_2, df_3, df_4, df_5)
```
